```python
import os
from glob import glob
import reciprocalspaceship as rs
```

```python
base_dir = "/Users/tjlane/Desktop/PL-workshop/dark_stability/"
column_name = "IMEAN"

df = None

mtz_paths = sorted(glob(base_dir + "dark_intensities/*.mtz"))
for i,mtz_path in enumerate(mtz_paths):
    print(mtz_path)

    if df is None:
        df = rs.read_mtz(mtz_path)[[column_name]]
    else:
        ds_name = os.path.basename(mtz_path)[:-4]
        new_df = rs.read_mtz(mtz_path)[[column_name]]
        df = df.join(new_df, how='inner', check_isomorphous=False, rsuffix=ds_name)
```

```python
mapper = {}
for column in df.columns:
    mapper.update({column: column[8:]})
mapper['F'] = 'dark'
print(mapper)

df = df.rename(columns=mapper)

df
```

```python
corr = df.corr()
html = corr.style.background_gradient(cmap='coolwarm_r', axis=None).set_precision(3)
html
```

```python
with open(base_dir + 'corrmatrix.html', 'w') as f:
    f.write(html.render())
```
