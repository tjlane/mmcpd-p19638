```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
model_vs_data = np.loadtxt('./rfactors-combi10ns.txt', skiprows=2)
no_solvent    = np.loadtxt('./rfactors-combi10ns-no-solvent.txt', skiprows=2)
```

```python
plt.figure()

ax = plt.subplot(211)
ax.plot(model_vs_data[:,0], model_vs_data[:,1:], 'o-')
ax.legend(['r-work', 'r-free'])
ax.set_ylabel('r factor')

ax = plt.subplot(212)
ax.plot(no_solvent[:,0], no_solvent[:,1:], 'o-')
ax.set_ylabel('r factor')
ax.set_xlabel('% dark')

plt.tight_layout()
plt.show()
```
